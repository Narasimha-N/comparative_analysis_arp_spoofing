```python
import pandas as pd
import os

ciciot_attack_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\CICIoT2023_Attack_Dataset.csv"
ciciot_benign_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\CICIoT2023_Benign_Dataset.csv"

iot_nid_attack_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\IoT_Network_Intrusion_Attack_Dataset.csv"
iot_nid_benign_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\IoT_Network_Intrusion_Benign_Dataset.csv"

uqiot_attack_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\UQ_IoT_IDS_Attack_Dataset.csv"
uqiot_benign_csv=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Kitsune-py-master\Extracted 20k featured CSV files\UQ_IOT_IDS_Benign_Dataset.csv"

output_dir=r"C:\Users\NARASIMHA N\OneDrive\Desktop\Kitsune Feature Extraction\Dataset Preprocessing\Remove Uncommon Features"
print("Paths are specified")
```

```python
#read pcaps and print the numbe of features of all datasets

ciciot_attack=pd.read_csv(ciciot_attack_csv)
ciciot_benign=pd.read_csv(ciciot_benign_csv)

iot_nid_attack=pd.read_csv(iot_nid_attack_csv)
iot_nid_benign=pd.read_csv(iot_nid_benign_csv)

uqiot_attack=pd.read_csv(uqiot_attack_csv)
uqiot_benign=pd.read_csv(uqiot_benign_csv)

print(f"CICIoT2023 Attack Dataset: {ciciot_attack.shape}")
print(f"CICIoT2023 Benign Dataset: {ciciot_benign.shape}")

print(f"IoT Network Intrusion Attack Dataset: {iot_nid_attack.shape}")
print(f"IoT Network Intrusion Benign Dataset: {iot_nid_benign.shape}")

print(f"UQ-IoT-IDS Attack Dataset: {uqiot_attack.shape}")
print(f"UQ-IoT-IDS Benign Dataset: {uqiot_benign.shape}")
```

```python
#find and remove uncommon features across attack and benign datasets of same dataset

ciciot_common_features= [feature for feature in ciciot_attack.columns if feature in set(ciciot_benign.columns)]
iot_nid_common_features= [feature for feature in iot_nid_attack.columns if feature in set(iot_nid_benign.columns)]
uqiot_common_features= [feature for feature in uqiot_attack.columns if feature in set(uqiot_benign.columns)]

print(f" CICIoT2023 Common features: {len(ciciot_common_features)}")
print(f"IoT-Network Intrusion common features: {len(iot_nid_common_features)}")
print(f"UQ-IoT-IDS common features: {len(uqiot_common_features)}")

ciciot_attack=ciciot_attack[ciciot_common_features]
ciciot_benign=ciciot_benign[ciciot_common_features]

iot_nid_attack=iot_nid_attack[iot_nid_common_features]
iot_nid_benign=iot_nid_benign[iot_nid_common_features]

uqiot_attack=uqiot_attack[uqiot_common_features]
uqiot_benign=uqiot_benign[uqiot_common_features]

print("After assigning the common features for all attack and benign datasets:")
print(f"CICIoT2023 Attack: {ciciot_attack.shape}")
print(f"CICIoT2023 Benign: {ciciot_benign.shape}")
print(f"IoT Network Intrusion Attack Dataset: {iot_nid_attack.shape}")
print(f"IoT Network Benign Dataset: {iot_nid_benign.shape}")
print(f"UQ-IoT-IDS Attack Dataset: {uqiot_attack.shape}")
print(f"UQ-IoT-IDS Benign Dataset: {uqiot_benign.shape}")
```

```python
#find and remove uncommon features across different datasets 

ciciot_features=set(ciciot_attack.columns)
iot_nid_features=set(iot_nid_attack.columns)
uqiot_features=set(uqiot_attack.columns)

cross_dataset_common_features= ciciot_features & iot_nid_features & uqiot_features

cross_dataset_common_ordered= [feature for feature in ciciot_attack.columns if feature in cross_dataset_common_features]

print(f"Features after cross-dataset alignment: {len(cross_dataset_common_ordered)}")

dropped_features_from_ciciot= ciciot_features - cross_dataset_common_features
dropped_features_from_iot_nid= iot_nid_features - cross_dataset_common_features
dropped_features_from_uqiot= uqiot_features - cross_dataset_common_features

print(f"\nDropped features from CICIoT2023 dataset: {sorted(dropped_features_from_ciciot)}")
print(f"\nDropped features from IoT-NID dataset: {sorted(dropped_features_from_iot_nid)}")
print(f"\nDropped features from UQ-IoT dataset: {sorted(dropped_features_from_uqiot)}")
```

```python
#Apply cross dataset common features to all datasets

ciciot_attack=ciciot_attack[cross_dataset_common_ordered]
ciciot_benign=ciciot_benign[cross_dataset_common_ordered]

iot_nid_attack=iot_nid_attack[cross_dataset_common_ordered]
iot_nid_benign=iot_nid_benign[cross_dataset_common_ordered]

uqiot_attack=uqiot_attack[cross_dataset_common_ordered]
uqiot_benign=uqiot_benign[cross_dataset_common_ordered]

print("Final shapes after all datasets having the same number of features:")
print(f"CICIoT2023 Attack: {ciciot_attack.shape}")
print(f"CICIoT2023 Benign: {ciciot_benign.shape}")
print(f"IoT Network Intrusion Attack Dataset: {iot_nid_attack.shape}")
print(f"IoT Network Benign Dataset: {iot_nid_benign.shape}")
print(f"UQ-IoT-IDS Attack Dataset: {uqiot_attack.shape}")
print(f"UQ-IoT-IDS Benign Dataset: {uqiot_benign.shape}")

print(f"All columns are identical: {(list(ciciot_attack.columns)==list(ciciot_benign.columns)==list(iot_nid_attack.columns)==list(iot_nid_benign.columns)==list(uqiot_attack.columns)==list(uqiot_benign.columns))}")

print(f"Features: {ciciot_attack.columns}")
```

```python
#save all CSVs having the same feature sets

ciciot_attack.to_csv(os.path.join(output_dir,"CICIoT2023_Attack.csv"), index=False)
ciciot_benign.to_csv(os.path.join(output_dir,"CICIoT2023_Benign.csv"), index=False)

iot_nid_attack.to_csv(os.path.join(output_dir,"IoT_NID_Attack.csv"), index=False)
iot_nid_benign.to_csv(os.path.join(output_dir,"IoT_NID_Benign.csv"), index=False)

uqiot_attack.to_csv(os.path.join(output_dir,"UQIoT_Attack.csv"), index=False)
uqiot_benign.to_csv(os.path.join(output_dir,"UQIoT_Benign.csv"), index=False)

print(f"All CSV files are saved in this folder: {output_dir}")
```
